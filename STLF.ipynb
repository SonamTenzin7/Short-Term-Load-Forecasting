# Adding new features to the existing dataframe to match the Bhutan-specific variables

# Electricity Demand in Bhutan is Driven by

* Hydropwer generation patterns(seasonal)
* Weather conditions (temperature, rainfall)
* Daily and Weekly usage cycles (peak vs off-peak hours)
* Residential load
* Holiday has a smaller impact


```python
# More realistic hydropower features for Bhutan
np.random.seed(42)

# 1. Rainfall (monsoon pattern - higher June-Sept)
monsoon_months = [6,7,8,9]
df['rainfall'] = np.where(
    df.index.month.isin(monsoon_months),
    np.random.uniform(5, 15, size=len(df)),
    np.random.uniform(0, 5, size=len(df))
)

# 2. Reservoir levels (inverse of rainfall with lag)
df['reservoir_level'] = 100 - (df['rainfall'].rolling(7).mean() * 0.8).clip(0,40)

# 3. River flow (correlated with rainfall)
df['river_flow'] = df['rainfall'] * 10 + np.random.uniform(20,40, size=len(df))

# 4. Plants operating (correlated with demand)
df['hydropower_plants_operating'] = np.where(
    df['hour'].between(7,22),
    np.random.choice([2,3], size=len(df)),
    np.random.choice([1,2], size=len(df)))

# 5. Hydropower generation (seasonal pattern)
df['hydropower_generation'] = (
    0.4 * df['river_flow'] +
    0.3 * (100 - df['reservoir_level']) +  # lower reservoir = higher generation
    0.2 * df['rainfall'] +
    5 * df['hydropower_plants_operating'] +
    np.random.normal(0, 5, size=len(df))
)
```

```python
# Hydropower-weather interactions
df['hydro_temp_interaction'] = df['hydropower_generation'] * df['T2M_dav']

# Reservoir level change rate
df['reservoir_change_rate'] = df['reservoir_level'].diff()

# Monsoon season flag
df['is_monsoon'] = df.index.month.isin([6,7,8,9]).astype(int)
```

```python
# Bhutan-specific time features
df['is_peak_season'] = df.index.month.isin([10,11,3,4]).astype(int)  # Tourism peaks
df['is_offpeak_hours'] = df['hour'].isin([0,1,2,3,4,5]).astype(int)  # Late night
```

Essential Features to Keep:
- Temporal: hour, is_weekend, day_of_week, month
- Weather: T2M_day, T2M_loc
- Calendar: Holiday_ID, holiday
- New Hydropower: hydropower_generation, rainfall, reservoir_level
- New Interactions: hydro_temp_interaction, is_monsoon

Features to Monitor:
- hydropower_plants_operating (may need one-hot encoding)
- river_flow (check correlation with generation)
- reservoir_change_rate

# Feature engineering
1. lagged demand features
  - Demand often shows strong temporal patterns
  - Add lagged values like previous hour, same hour yesterday, etc

2. Rolling window statistics
  - Rolling mean/mac/min over past few hours to capture trends

3. Temperature-Based Features
  - Hourly change in temperature
  - Temperature bins or categories (cold, mild, hot)

4. Time-Based interactions
  - Hour x is_weekend
  - Hour x is_monsoon
  - Day of week x holiday flag

5. Cyclic Time features
   - Encode hour,day_of_week, month using sine/cosine to preserve cyclic nature

```python
# Lag features (ensure no leakage in final modeling)
df['demand_lag_1'] = df['nat_demand'].shift(1)     # previous hour
df['demand_lag_24'] = df['nat_demand'].shift(24)   # same hour yesterday
df['demand_lag_168'] = df['nat_demand'].shift(168) # same hour last week

# Optional: rolling average of demand
df['demand_rolling_mean_6h'] = df['nat_demand'].rolling(6).mean()
df['demand_rolling_mean_24h'] = df['nat_demand'].rolling(24).mean()
```

```python
# Rolling mean of temperature (e.g., Dav temperature)
df['T2M_dav_rolling_mean_6h'] = df['T2M_dav'].rolling(6).mean()
df['T2M_dav_rolling_mean_24h'] = df['T2M_dav'].rolling(24).mean()

# Max temp in past 6/24 hours
df['T2M_dav_rolling_max_6h'] = df['T2M_dav'].rolling(6).max()
```

```python
# Hourly temperature change
df['temp_change_hourly'] = df['T2M_dav'].diff()

# Absolute rate of change
df['abs_temp_change'] = df['temp_change_hourly'].abs()
```

```python
import numpy as np

def create_cyclic_features(df, col, max_val):
    df[col + '_sin'] = np.sin(2 * np.pi * df[col]/max_val)
    df[col + '_cos'] = np.cos(2 * np.pi * df[col]/max_val)
    return df

# Hour (023)
df = create_cyclic_features(df, 'hour', 24)

# Day of week (0=Monday to 6=Sunday)
df = create_cyclic_features(df, 'day_of_week', 7)

# Month (112)
df = create_cyclic_features(df, 'month', 12)
```

```python
# Hour  Weekend
df['hour_x_weekend'] = df['hour'] * df['is_weekend']

# Hour  Monsoon
df['hour_x_monsoon'] = df['hour'] * df['is_monsoon']

# Holiday  Hour
df['hour_x_holiday'] = df['hour'] * df['holiday']
```

# Step 1: Create a Function to Generate Features from Input Data

```python
import pandas as pd
import numpy as np
from datetime import datetime

def create_real_time_features(df):
    """
    Takes a DataFrame with at least 'datetime' and required variables,
    returns a DataFrame with all engineered features.
    """

    # Set datetime as index
    df = df.set_index('datetime')

    # Extract time-based features
    df['hour'] = df.index.hour
    df['day_of_week'] = df.index.dayofweek  # Monday=0, Sunday=6
    df['month'] = df.index.month
    df['is_weekend'] = df.index.weekday >= 5

    # Cyclical encoding of time
    def create_cyclic_features(data, col, max_val):
        data[col + '_sin'] = np.sin(2 * np.pi * data[col]/max_val)
        data[col + '_cos'] = np.cos(2 * np.pi * data[col]/max_val)
        return data

    df = create_cyclic_features(df, 'hour', 24)
    df = create_cyclic_features(df, 'day_of_week', 7)
    df = create_cyclic_features(df, 'month', 12)

    # Interaction features
    df['hour_x_weekend'] = df['hour'] * df['is_weekend']

    # Monsoon season flag (JuneSeptember)
    df['is_monsoon'] = df.index.month.isin([6,7,8,9]).astype(int)

    # Peak season (Tourism peak months - adjust based on Bhutan)
    df['is_peak_season'] = df.index.month.isin([3,4,10,11]).astype(int)

    # Off-peak hours (late night)
    df['is_offpeak_hours'] = df['hour'].isin([0,1,2,3,4,5]).astype(int)

    # Hydropower-related features (example logic  adjust accordingly)
    df['hydropower_plants_operating'] = np.where(df['hour'].between(7, 22), 2, 1)

    df['hydro_temp_interaction'] = df['T2M_dav'] * df['hydropower_plants_operating']

    # Example reservoir level (you can replace this with actual data)
    df['reservoir_level'] = 70  # placeholder; could be dynamic

    # Lagged features require historical data, so skip or simulate if needed
    # Rolling features also require history

    return df
```

```python
# Prepare new data
from datetime import datetime
import pandas as pd
import numpy as np

def create_real_time_features(timestamp=None):
    if timestamp is None:
        timestamp = datetime.now()

    # Create DataFrame with one row
    df = pd.DataFrame({'datetime': [timestamp]})

    # Extract basic time-based features
    df['hour'] = df['datetime'].dt.hour
    df['day_of_week'] = df['datetime'].dt.weekday  # Monday=0 ... Sunday=6
    df['month'] = df['datetime'].dt.month
    df['is_weekend'] = (df['datetime'].dt.weekday >= 5).astype(int)

    # Monsoon, peak season, off-peak hour flags
    df['is_monsoon'] = df['month'].isin([6,7,8,9]).astype(int)
    df['is_peak_season'] = df['month'].isin([3,4,10,11]).astype(int)
    df['is_offpeak_hours'] = df['hour'].isin([0,1,2,3,4,5]).astype(int)

    # Simulated weather data  replace with real API or forecast values
    df['T2M_dav'] = 18.5   # temperature at Dav
    df['T2M_toc'] = 20.0   # temperature at Toc

    # Calendar
    df['Holiday_ID'] = 0
    df['holiday'] = 0

    # Hydropower-related
    df['hydropower_plants_operating'] = 2  # adjust based on schedule

    # Cyclical encoding
    def create_cyclic_features(df, col, max_val):
        df[col + '_sin'] = np.sin(2 * np.pi * df[col]/max_val)
        df[col + '_cos'] = np.cos(2 * np.pi * df[col]/max_val)
        return df

    df = create_cyclic_features(df, 'hour', 24)
    df = create_cyclic_features(df, 'day_of_week', 7)
    df = create_cyclic_features(df, 'month', 12)

    # Interaction feature
    df['hydro_temp_interaction'] = df['T2M_dav'] * df['hydropower_plants_operating']

    return df
```
